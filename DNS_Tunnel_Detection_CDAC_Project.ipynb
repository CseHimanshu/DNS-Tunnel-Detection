```python
import ipaddress

# Function to convert IP address to integer
def ip_to_int(ip_address_str):
    ip = ipaddress.ip_address(ip_address_str)
    return int(ip)

# Apply the conversion function to the "SourceIP" column
data['SourceIP'] = data['SourceIP'].apply(ip_to_int)
data['DestinationIP'] = data['DestinationIP'].apply(ip_to_int)
# Display the updated DataFrame with numerical representation of SourceIP
print(data)
```

```python
data.drop("TimeStamp",axis=1, inplace=True)
```

```python
data['Tunneled'] = data['Tunneled'].map({'Yes':1 ,'No':0})
```

# Splitting


```python
X = data.drop("Tunneled", axis =1)
y = data["Tunneled"]
```

```python
sns.heatmap(X.corr())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
clf.score(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
accuracy_score(y_test,y_preds)
```

```python
#Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_scaled, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_scaled, y_train_scaled)
```

```python
model.score(X_test_scaled, y_test_scaled)
```

```python
y_preds_scaled = model.predict(X_test_scaled)
y_preds_scaled
```

```python
print(classification_report(y_test_scaled,y_preds_scaled))
```
